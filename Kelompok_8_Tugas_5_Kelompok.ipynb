```python
# Import Library yang Dibutuhkan
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Scraping Data
wikipedia = requests.get('https://simple.wikipedia.org/wiki/List_of_countries_by_continents', headers={'User-Agent': 'Mozilla/5.0'})
wiki_parsed = BeautifulSoup(wikipedia.content, 'html.parser')

wiki_parsed
```

```python
wikipedia.text
```

### Parsing

```python
# Nama Benua
benua = wiki_parsed.select('h2 span.mw-headline')
benua
```

```python
unwanted_words = ['Notes', 'References','Other website']
target_benua = [benua.text for benua in benua if benua.text not in unwanted_words]

target_benua
```

```python
# Benua Africa sampai North America
continents = wiki_parsed.find_all('table',  class_ = 'wikitable sortable')
continents
```

```python
# Benua Africa
hasil_tes1 = []
for row in continents[0].find_all('tr')[1:]:  # Skip the first row (header)
    columns = row.find_all('td')
    country = columns[2].text.strip()
    capital = columns[3].text.strip()
    population = columns[4].text.strip()
    area = columns[5].text.strip()
    currency = columns[6].text.strip()
    image_url = columns[1].find('img')['src']
    hasil_tes1.append({'negara':country, 'ibukota':capital, 'populasi':population, 'mata uang':currency, 'luas':area, 'bendera':image_url})

pd.DataFrame(hasil_tes1)
```

```python
#import module & preprocessing
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By

url = 'https://simple.wikipedia.org/wiki/List_of_countries_by_continents'
service = Service(executable_path=r'/usr/bin/chromedriver')
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(service=service, options=options)
```

```python
#web source
driver.get(url)
```

```python
#afrika, asia, eropa, amerika utara
df_negara = pd.DataFrame(columns=['nama','ibukota','populasi','luas','uang'])

nama = driver.find_elements('xpath','//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
ibukota = driver.find_elements('xpath','//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
populasi = driver.find_elements('xpath','//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
luas = driver.find_elements('xpath','//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[6]')
uang = driver.find_elements('xpath','//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[7]')

for i in range(len(nama)):
  df_negara = df_negara.append({'nama': nama[i].text, 'ibukota': ibukota[i].text, 'populasi': populasi[i].text,
                                'luas': luas[i].text, 'uang': uang[i].text}, ignore_index = True)
```

```python
df_negara
```

```python
#amerika selatan, oceiana
df_negara1 = pd.DataFrame(columns=['nama','ibukota','populasi','luas','uang'])

nama1 = driver.find_elements('xpath','//table[@class="wikitable"]/tbody/tr/td[2]')
ibukota1 = driver.find_elements('xpath','//table[@class="wikitable"]/tbody/tr/td[3]')
populasi1 = driver.find_elements('xpath','//table[@class="wikitable"]/tbody/tr/td[4]')
luas1 = driver.find_elements('xpath','//table[@class="wikitable"]/tbody/tr/td[5]')
uang1 = driver.find_elements('xpath','//table[@class="wikitable"]/tbody/tr/td[7]')

for i in range(len(nama1)):
  df_negara1 = df_negara1.append({'nama': nama1[i].text, 'ibukota': ibukota1[i].text, 'populasi': populasi1[i].text,
                                'luas': luas1[i].text, 'uang': uang1[i].text}, ignore_index = True)
```

```python
#merge data
df = pd.concat([df_negara, df_negara1], axis=0)
```
