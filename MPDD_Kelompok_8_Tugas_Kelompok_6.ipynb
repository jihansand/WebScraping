Akan diambil data keberangkatan kereta api dari Jakarta melalui stasiun Gambir (GMR) dan Pasar Senen (PSE) menuju stasiun Semarang Tawang (SMT) tanggal 30 Maret 2024.

```python
list_stasiun = ['GMR', 'PSE']

list_df = []

# Iterasi stasiun
for stasiun in list_stasiun:
    # Requests API
    url = f"https://www.tiket.com/ms-gateway/tix-train-search-v2/journeys"
    params = {
    "orig": stasiun,
    "otype": "STATION",
    "dest": "SMT",
    "dtype": "STATION",
    "ttype": "ONE_WAY",
    "ddate": "20240330",
    "acount": "1",
    "icount": "0"
    }

    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}

    response = requests.request("GET", url, params=params, headers=headers)

    # ubah JSON
    json_data = json.loads(response.text)['data']['departJourneys']['journeys']
    segment_schedules = []
    for journey in json_data:
      segment_schedules.extend(journey.get('segmentSchedules', []))

    # ubah menjadi data frame
    df = pd.json_normalize(segment_schedules)

    print(f'Keberangkatan dari {stasiun} sebanyak {len(df)}')

    # input df hasil req ke list
    list_df.append(df)

# gabungkan df dari stasiun yg diminta
df_hasil = pd.concat(list_df, ignore_index=True)
df_hasil
```

Akan diambil data keberangkatan kereta api dari Jakarta melalui stasiun Gambir (GMR) dan Pasar Senen menuju stasiun Semarang Tawang (SMT) tanggal 30 Maret sampai 1 April 2024.

```python
list_stasiun = ['GMR', 'PSE']
tanggal = ['20240330','20240331','20240401']

list_df = []

# Iterasi stasiun
for stasiun in list_stasiun:
    df_stasiun = pd.DataFrame()
    for tgl in tanggal:
    # Requests API
      url = f"https://www.tiket.com/ms-gateway/tix-train-search-v2/journeys"
      params = {
          "orig": stasiun,
          "otype": "STATION",
          "dest": "SMT",
          "dtype": "STATION",
          "ttype": "ONE_WAY",
          "ddate": tgl,
          "acount": "1",
          "icount": "0"
          }

      headers = {
          "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}

      response = requests.request("GET", url, params=params, headers=headers)

      # ubah JSON
      json_data = json.loads(response.text)['data']['departJourneys']['journeys']
      segment_schedules = []
      for journey in json_data:
        segment_schedules.extend(journey.get('segmentSchedules', []))

      # ubah menjadi data frame
      df = pd.json_normalize(segment_schedules)

      print(f'Keberangkatan dari {stasiun} pada {tgl} sebanyak {len(df)}')

      # Tambahkan kolom tanggal ke dataframe
      df['Tanggal'] = tgl

      # gabungkan dataframe ke dataframe stasiun
      df_stasiun = pd.concat([df_stasiun, df], ignore_index=True)

    # input df hasil req ke list
    list_df.append(df_stasiun)

# gabungkan df dari berbagai tanggal dan stasiun yang diminta
df_hasil = pd.concat(list_df, ignore_index=True)
df_hasil[['trainName', 'departureDate', 'departureTime', 'arrivalDate', 'arrivalTime',
                'availableSeats', 'departureStation.code',
                'arrivalStation.code', 'wagonClass.detail', 'subClass.code']]
```
